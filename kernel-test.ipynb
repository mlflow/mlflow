```python
import mlflow

with mlflow.start_run():
    import numpy as np
    from sklearn.linear_model import LinearRegression

    X = np.array([[1], [2], [3], [4]])
    y = np.array([2, 4, 6, 8])

    model = LinearRegression()
    model.fit(X, y)

    mlflow.sklearn.log_model(model, "simple_linear_regression_model")
```
